```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

```python
data2 = pd.read_csv('Lab2_dataframe.csv')
```

```python
data2.head(5)
```

```python
#Get the numeric data into dataframe called numerical and categorical columns in a dataframe called categoricals.

categoricals =  data2.select_dtypes('object')
numericals = data2.select_dtypes(include=['float64', 'int64'])
```

```python
categoricals.head(3)
```

```python
numericals.head(2)
```

```python
#Use seaborn library to construct distribution plots for the numerical variables

sns.pairplot(numericals)
plt.show()
```

```python
for col in numericals.columns:
    sns.displot(numericals[col])
    plt.show()
```

```python
#Use Matplotlib to construct histograms

for col in numericals.columns:
    plt.hist(numericals[col], bins=20)
    plt.show()
```

```python
# Do the distributions for different numerical variables look like a normal distribution
#Analysing all the distribution variables
# None of them are a normal distribution

for col in numericals.columns:
    sns.displot(numericals[col], kde=True)
plt.show()
```

```python
#For the numerical variables, check the multicollinearity between the features. 
#Please note that we will use the column total_claim_amount later as the target variable.

sns.heatmap(numericals.corr(),annot=True, annot_kws={"size": 8})
plt.show()
```
